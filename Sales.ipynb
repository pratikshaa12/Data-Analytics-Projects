###### 2. Most Selling product?

b.By  Revenue

```python
df.head()
```

```python
df.groupby('Product')['Quantity'].sum()
```

```python
df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
most_quantity= df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
most_quantity
```

```python
type(most_quantity)
```

```python
most_quantity=most_quantity.reset_index()
most_quantity
```

```python
type(most_quantity)
```

```python
plt.figure(figsize= (9,4))
plt.bar(most_quantity['Product'], most_quantity['Quantity'], color=['red', 'black', 'green', 'yellow','cyan'], width=0.4)
plt.title=('Most selling Product- By Quantity')
plt.xlabel('Product')
plt.ylabel('Quantity');
```

```python
# By Revenue
df.head()
```

```python
df['Revenue']= df['Price']*df['Quantity']    # to create new column
```

```python
df
```

```python
df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
```

```python
most_revenue=df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
most_revenue
```

```python
most_revenue=most_revenue.reset_index()   # to convert series into a dataframe
most_revenue
```

```python
plt.figure(figsize= (9,4))
plt.bar(most_revenue['Product'], most_revenue['Revenue'], color=['red', 'black', 'green', 'yellow','cyan'], width=0.3)
plt.title=('Most selling Product- By Revenue')
plt.xlabel('Product')
plt.ylabel('Revenue');
```
